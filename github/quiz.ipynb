**1. GitHub**

1.1. Что такое GitHub?

    Облачная платформа для хостинга Git-репозиториев и совместной разработки ПО
    с инструментами для обсуждений, тестирования и управления проектами.

1.2. Как GitHub связан с Git?

    Git система контроля версий локально на компьютере.
    GitHub — удалённый хостинг для Git-репозиториев с дополнительными функциями совместной работы.

1.3. Чем отличается fork репозитория от его клонирования (clone)?

    Fork создаёт копию репозитория на GitHub под вашей учётной записью.
    Clone скачивает репозиторий (оригинал или форк) на ваш компьютер для локальной работы.

1.4. Зачем нужны и как работают pull requests?

    Pull request — запрос на слияние изменений из вашей ветки/форка
    в основной репозиторий. Позволяет обсудить, проверить и принять изменения.

1.5. GitHub использует ваш почтовый адрес для привязки ваших Git коммитов к вашей учётной записи?

    Да. GitHub привязывает коммиты к учётной записи по email-адресу,
    указанному в настройках коммита (git config user.email).

1.6 Какая команда генерирует SSH ключ для Доступа по SSH к репозиторию (Рисунок 83)

    ssh-keygen -t ed25519 -C "your_email@example.com" (или -t rsa -b 4096 для RSA).
