## Список вопросов и ответов по Cpython 


#### Введение в Python и CPython 
1. Что такое CPython и чем он отличается от Python? 

Cpython это одна из реализаций языка Python. По совместительству он является и интерпретатором. 

2. Сколько существует реализаций Python, и какая из них самая популярная? 

Всего существует 6 реализаций. Самая популярная - Cpython. 

3. На каком языке написан CPython? 

На си.

#### Запуск файла с помощью CPython 
10. Какая команда используется для запуска файла с помощью CPython? 

python file.py 

11. Можно ли запускать текстовые файлы через интерпретатор Python? Почему? 

Можно, так как CPython не проверяет расширение до тех пор, пока оно содержит корректный код. Он по прежнему будет считывать и пытаться интерпретировать строки как инструкции, которые ему нужно выполнить.

#### Введение в PyPy 
13. Чем PyPy отличается от CPython? 

PyPy еще быстрее, чем cpython. Но так как это достаточно новый интерпретатор, он может быть не совместим с какими-то проектами на питоне. 

14. Почему PyPy не может использоваться для всех проектов на Python? 

Так как он может быть несовместим с ними. Некоторые библиотеки (например numpy) работают корректно только с cpython. 

15. Где можно скачать PyPy? 

На официальном сайте https://pypy.org

#### Установка и запуск PyPy 
17. Как установить PyPy после скачивания? 

После скачивания нужно переместить архив в удобную для нас папку и разорхивировать его так. Внутри будет находиться сам интерпретатор. Добавить в системную переменную PATH путь до него. 

18. Как запустить файл с помощью PyPy? 

pypy file.py

19. Почему PyPy выполняет код быстрее, чем CPython? 

 - PyPy использует JIT-компиляцию, которая позволяет динамически компилировать часто исполняемый байт-код в машинный код. Это значительно ускоряет выполнение повторяющихся операций. 
 - PyPy оптимизирует работу интерпретатора на уровне интерпретации, что позволяет снизить накладные расходы и увеличить скорость выполнения кода. 
 - PyPy использует более эффективный сборщик мусора и аллокаторы памяти.
