И либо жмём accept current changes - тогда примутся изменения с интернета.  Либо accept incoming changes - тогда будут наши (нежелательно, жмём в редких случаях). Либо accept both changes - оба изменения. Либо ручками удаляем то что не нужно и жмём - resolve in merge editor
