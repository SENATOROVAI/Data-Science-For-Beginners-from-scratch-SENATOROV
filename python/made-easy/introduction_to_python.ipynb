### Ответьте на вопросы
1. Python - это программное обеспечение с открытым исходным кодом. Это то же самое, что и бесплатное ПО?
Open Source означает открытость исходного кода для просмотра и модификации.

2. У всех ли бесплатных программ открытый исходный код? А если нет, то в чем разница?
Не у всех. Есть бесплатные, но закрытые.

3. Python поддерживает динамическую типизацию. Что это такое?
Тип переменной определяется во время выполнения, а не при написании кода

4. Назовите 5 самых популярных языков программирования для специалистов по анализу данных.


5. В чем заключается преимущество Python по сравнению с языком С?
- Простота синтаксиса — быстрее писать код
- Автоматическое управление памятью
- Богатые библиотеки для разных задач
- Интерактивность — можно тестировать код сразу
- Кроссплатформенность без перекомпиляции

6. Python портативен. Что в этом контексте означает «портативность»?
код работает на Windows, macOS, Linux без изменений, так как Python — интерпретируемый язык.

7. В чем разница между «расширяемым» и «встраиваемым» языком?
Расширяемый: можно добавлять модули на C/C++ для ускорения
Встраиваемый: Python можно интегрировать в другие приложения как скриптовый движок

8. В чем смысл IDE? Чем она отличается от командной строки?
- IDE — полная среда разработки с редактором, отладчиком, автодополнением в одном окне.
- Командная строка — текстовый интерфейс для выполнения команд по одной.

9. Как открыть существующий документ Jupyter Notebook? Чем эта процедура отличается от открытия РDF-файла или текстового файла?
- Jupyter: jupyter notebook filename.ipynb -> открывается в браузере как интерактивная среда.
- PDF/текст: двойной клик → открывается в программе просмотра статично.

10. В чем разница между «ячейками разметки Markdown и ячейками кода в JN. Для чего они нужны?
- Markdown ячейки: текст, заголовки, формулы для документации
- Code ячейки: исполняемый Python-код с выводом результатов
