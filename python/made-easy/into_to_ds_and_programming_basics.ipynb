Data science helps in decision-making, pattern discovery, predictive analysis, and data visualization. It allows us to:

- Identify key questions and root causes of problems.
- Find patterns in raw data.
- Build predictive models.
- Visualize insights using graphs and dashboards.
- Enable machine intelligence.
- Analyze customer sentiment and improve recommendations.
- Enhance business growth by making faster and better decisions.

**Components of Data Science**
- Data Mining
- Data Analytics
- Data Engineering
- Visualization
- Statistical Analysis

- Artificial Intelligence (AI) focuses on creating machines that mimic human behavior. It dates back to Alan Turing's early work in 1936 but still cannot fully replace humans.
- Machine Learning (ML) extracts knowledge from data, using supervised (with human input) or unsupervised (without human input) training methods.
- Deep Learning (DL) uses multi-layer neural networks for complex tasks where traditional ML is insufficient.
- Big Data deals with massive, often unstructured data, requiring specialized tools for high-performance processing.

**A data scientist extracts valuable insights from business data by following these steps:**

- Identify key questions to understand the problem.
- Collect data from various sources (enterprise, public, etc.).
- Process and clean data for analysis.
- Apply machine learning or statistical models for insights.
- Present findings to stakeholders for decision-making.


**To succeed in data science, the following skills are essential:**

- Programming: Knowledge of Python or R is needed, with Python being the most popular due to its ease of use and vast libraries.
- Statistics: A strong foundation in statistics helps extract meaningful insights from data.
- Databases: Understanding how to manage and extract data from databases is crucial.
- Modeling: Mathematical models enable predictions and help identify the best algorithms for machine learning.

## Compiler

The compiler is a program that converts the source code to the machine code.

## Interpreter

In computer programming, an interpreter is a computer program that
directly executes instructions written in a programming language,
without requiring them previously to have been compiled into a
machine language program


## Comparison: Compiler vs. Interpreter

- Compiler: Translates entire code at once.
- Interpreter: Executes code line by line.
- Compiler: Generates a stand-alone machine code program.
- Interpreter: Executes directly without creating a separate file.
- Compiler: Source code not required after compilation.
- Interpreter: Source code required every time for execution.
- Compiler: Faster (precompiled).
- Interpreter: Slower (translates during runtime).
