1) Python - это программное обеспечение с открытым исходным кодом.

разница.

распростронять по своему усмотрению нельзя или изменять
Пример: WinRar

Динамическая типизация - это когда тип переменной определяется сам
в ходе исполнения программы, а не на этапе компиляции

Встраиваемый — можно встроить сам язык в другую программу.
