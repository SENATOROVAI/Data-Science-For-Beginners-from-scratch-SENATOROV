## 2.12. Упражнения
### 2.12.1. Ответьте на вопросы
1. Python - это программное обеспечение с открытым исходным кодом. Это то же
самое, что и бесплатное ПО?
Ответ: нет, потому у бесплатного ПО может быть закрытый исходный код, а
также всё зависит от лицензии
2. У всех ли бесплатных программ открытый исходный код? А если нет, то в чем
разница?
Ответ: нет, т.к. разработчики бесплатного ПО могут не захотеть делиться
исходным кодом
3. Python поддерживает динамическую типизацию. Что это такое?
Ответ: интерпретатор автоматически определяет, какого типа значение
присваивается переменной. Предварительно не нужно объявлять тип переменной.
4. Назовите 5 самых популярных языков программирования для специалистов по
анализу данных.
Ответ: python, R, SQL, C++, java, matlab
5. В чем заключается преимущество Python по сравнению с языком С?
Ответ: портабельность и кроссплатформенность, простота, интерпретируемость,
автоматическое управление памятью
6. Python портативен. Что в этом контексте означает «портативность»?
Ответ: код работает на разных устройствах и ОС правильно
7. В чем разница между «расширяемым» и «встраиваемым» языком?
Ответ:
- Расширяемый язык позволяет писать модули на других языках и подключать их
к python для улучшения производительности или доступа к низкоуровневым
функциям (например, ручное управление памятью на плюсах).
- Встраиваемый язык можно встроить в приложение, например, на C/C++. Т.е.
интерпретатор можно подключить к коду, работающему на других языках.
8. В чем смысл IDE? Чем она отличается от командной строки?
Ответ: подсветка синтаксиса, дополнительные функции, встроенный редактор
кода, терминал, файловая система и т.д.
9. Как открыть существующий документ Jupyter Notebook? Чем эта процедура
отличается от открытия РDF-файла или текстового файла?
Ответ: jupyter notebook name.ipynb. Тем, что ipynb - не статический файл, а
контейнер для кода и вывода, который открывается через веб-сервер Jupyter в
браузере. PDF и текстовый файл открываются в просмотрщике/редакторе как
статический документ, без возможности интерактивно выполнять вычисления и
изменять содержимое ячеек.
10. В чем разница между «ячейками разметки Markdown» и «ячейками кода»
в Jupyter Notebook? Для чего они нужны?
Ответ: Markdown  ячейки используются для текста, который имеет возможность к
форматированию. Обычно там поясняется написанный код или же математическое
обоснование к коду

5. Поищите информацию о волшебных командах в IPython и Jupyter Notebook.
Ответ:

- %timeit expr - автоматически запускает expr много раз, чтобы вывести
среднее время выполнения.
- %run script.py - запускает внешний Python-скрипт в текущем пространстве
имён.
- %load filename - загружает содержимое файла в текущую ячейку.
