#### Примеры итераторов

Возведение в квадрат

```python
class Square:
  def __init__(self, seq):
    self._seq = seq
    self._idx = 0

  def __iter__(self):
    return self

  def __next__(self):
    if self._idx < len(self._seq):
      square = self._seq[self._idx] ** 2
      self._idx += 1
      return square
    else:
      raise StopIteration
```

```python
square = Square([1,2,3,4,5])
square
```

```python
for s in square:
  print(s)
```

Счетчик

```python
class Counter:
  def __init__(self, start = 3, stop = 9):
    self._current = start - 1
    self._stop = stop

  def __iter__(self):
    return self

  def __next__(self):
    self._current += 1
    if self._current < self._stop:
      return self._current
    else:
      raise StopIteration
```

```python
counter = Counter()
counter
```

```python
print(next(counter))
print(next(counter))
```

```python
for c in counter:
  print(c)
```

Класс Iterator модуля collections.abc

```python
from collections.abc import Iterator

class Counter2(Iterator):
  def __init__(self, start = 3, stop = 9):
    self._current = start - 1
    self._stop = stop

  def __next__(self):
    self._current += 1
    if self._current < self._stop:
      return self._current
    else:
      raise StopIteration
```

```python
for c in Counter2():
  print(c)
```

Бесконечный итератор

```python
class FibIterator:
  def __init__(self):
    self._idx = 0
    self._current = 0
    self._next = 1

  def __iter__(self):
    return self

  def __next__(self):
    self._idx += 1
    self._current, self._next = (self._next, self._current + self._next)
    return self._current
```

```python
limit = 10

for f in FibIterator():
  print(f)
  limit -= 1
  if limit == 0:
    break
```

### Генератор

#### Простой пример

```python
def sequence(n):
  res = [x for x in range(1, n+1)]
  return res
```

```python
sequence(5)
```

```python
def sequence_gen(n):
  for x in range(1, n+1):
    yield x
```

```python
sequence_gen(5)
```

```python
seq_5 = sequence_gen(5)

print(next(seq_5))
print(next(seq_5))
```

```python
for i in seq_5:
  print(i)
```

### Модуль itertools

```python
from itertools import count, cycle, chain
```

#### Функция `count()`

```python
natural_numbers = count(start=1, step=0.5)

for num in natural_numbers:
    print(num)
    if num == 2:
      break
```

```python
list_ = ['A', 'B', 'C', 'D']
for i in zip(count(), list_):
  print(i)
```

```python
def f(x):
  return x**2+x-2

f_x = map(f, count())
next(f_x)
```

```python
for val in f_x:
  print(val)
  if val > 10:
    break
```

#### Функция `cycle()`

```python
list_ = [1,2,3]
iterator = cycle(list_)

limit = 5
for i in iterator:
  print(i)
  limit -= 1
  if limit == 0:
    break
```

```python
string = 'Python'
iterator = cycle(string)

limit = 10
for i in iterator:
  print(i)
  limit -= 1
  if limit == 0:
    break
```

```python
list(iterator)
```

```python
list(chain.from_iterable(['abc', 'def']))
```
