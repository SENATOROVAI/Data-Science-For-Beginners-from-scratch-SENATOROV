### Знакомство с декораторами

Простой декоратор

```python
def simple_decorator(func):
  def wrapper():
    print('Текст до вызова функции func().')
    func()
    print('Текст после вызова функции func().')
  return wrapper

def say_hello():
  print('Привет!')
```

```python
say_hello = simple_decorator(say_hello)
```

```python
say_hello()
```

Конструкция @decorator

```python
@simple_decorator
def say_hi():
  print('Снова, привет!')
```

```python
say_hi()
```

Функции с аргументами

```python
@simple_decorator
def say_hello_with_name(name):
  print(f'Привет, {name}!')
```

```python
# say_hello_with_name('Алексей')
```

```python
def decorator_with_name_argument(func):
  def wrapper(name):
    print('Текст до вызова функции func().')
    func(name)
    print('Текст после вызова функции func().')
  return wrapper
```

```python
@decorator_with_name_argument
def say_hello_with_name(name):
  print(f'Привет, {name}!')
```

```python
say_hello_with_name('Алексей')
```

```python
def decorator_with_arguments(func):
  def wrapper(*args, **kwargs):
    print('Текст до вызова функции func().')
    func(*args, **kwargs)
    print('Текст после вызова функции func().')
  return wrapper
```

```python
@decorator_with_arguments
def say_hello_with_argument(name):
  print(f'Привет, {name}!')
```

```python
say_hello_with_argument('Алексей')
```

```python
import functools

def repeat_twice(func):
  @functools.wraps(func)
  def wrapper(*args, **kwargs):
    func(*args, **kwargs)
    func(*args, **kwargs)
  return wrapper
```

```python
@repeat_twice
def square(x):
  """Squares a number"""
  print(x * x)
```

```python
square.__name__, square.__doc__
```

```python
square.__wrapped__
```

```python
def repeat_twice(func):
  def wrapper(*args, **kwargs):
    func(*args, **kwargs)
    func(*args, **kwargs)
  functools.update_wrapper(wrapper, func)
  return wrapper
```

```python
@repeat_twice
def power(x,n):
  """Raises to a power"""
  print(x ** n)
```

```python
power(2,3)
```

```python
power.__doc__
```

### Примеры декораторов

Создание логов

```python
def logging(func):
  def wrapper(*args, **kwargs):
    print(f'Calling {func.__name__} with args: {args}, kwargs: {kwargs}')
    result = func(*args, **kwargs)
    print(f'{func.__name__} returned: {result}')
    return result
  return wrapper
```

```python
@logging
def power(x,n):
  return x**n

power(5, 3)
```

Время исполнения функции

```python
import time

def timer(func):
  def wrapper(*args, **kwargs):
    start_time = time.time()
    result = func(*args, **kwargs)
    end_time = time.time()
    print(f"{func.__name__} executed in {end_time - start_time:.4f} seconds")
    return result
  return wrapper
```

```python
@timer
def delayed_function(t):
  time.sleep(t)
  return 'execution completed'

delayed_function(2)
```

Статические методы

```python
class CatClass:

  species = 'кошка'

  def __init__(self, color):
    self.color = color
    self.type_ = 'cat'

  def info(self):
    print(self.color, self.type_)

  @classmethod
  def get_species(cls):
    print(cls.species)
    # нет доступа к переменным color и type_

  @staticmethod
  def convert_to_pounds(x):
    print(f'{x} kg is approximately {x * 2.205} pounds')
    # нет доступа к переменным species, color и type_
```

```python
CatClass.convert_to_pounds(4)
```

```python
cat = CatClass('gray')
cat.convert_to_pounds(5)
```

```python
@timer
class CatClass:

  def __init__(self, color):
    self.color = color
    self.type_ = 'cat'

  def info(self):
    time.sleep(2)
    print(self.color, self.type_, sep=', ')
```

```python
cat.weight, getattr(cat, 'weight')
```

```python
def add_attribute(attribute_name, attribute_value):
  def wrapper(cls):
    setattr(cls, attribute_name, attribute_value)
    return cls
  return wrapper
```

```python
@add_attribute('species', 'кошка')
class CatClass:

  def __init__(self, color):
    self.color = color
    self.type_ = 'cat'
```

### Декораторы с аргументами

```python
def repeat(n_times):
  def inner_decorator(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
      for _ in range(n_times):
        func(*args, **kwargs)
    return wrapper
  return inner_decorator
```

```python
@repeat(n_times=3)
def say_hello(name):
  print(f'Привет, {name}!')
```

```python
say_hello('Алексей')
```
