19. Почему PyPy выполняет код быстрее, чем CPython?
PyPy часто быстрее CPython, потому что использует JIT-компилятор (Just-In-Time). В то время как CPython интерпретирует код построчно, PyPy анализирует часто используемые части кода и компилирует их напрямую в оптимизированный машинный код, который выполняется значительно быстрее.
